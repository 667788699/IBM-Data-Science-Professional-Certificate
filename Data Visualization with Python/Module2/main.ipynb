## Pie Chart

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
df.set_index('Country', inplace=True)

```

```python
years = list(map(str, range(1980, 2014)))

df_cotinent['Total'].plot(kind='pie')
```

## Boxplot

```python
df.loc[['China', 'India'], years].transpose().plot(kind='box')
```

```python
fig = plt.figure() # create figure

ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot). See tip below**

# Subplot 1: Box plot
df.loc[['China', 'India'], years].transpose().plot(kind='box', color='blue', vert=True, figsize=(20, 6), ax=ax0) # add to subplot 1
ax0.set_title('Box Plots of Immigrants from China and India (1980 - 2013)')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')

# Subplot 2: Line plot
df.loc[['China', 'India'], years].transpose().plot(kind='line', figsize=(20, 6), ax=ax1) # add to subplot 2
ax1.set_title ('Line Plots of Immigrants from China and India (1980 - 2013)')
ax1.set_ylabel('Number of Immigrants')
ax1.set_xlabel('Years')

plt.show()
```

## matplotlib

```python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv'

df = pd.read_csv(url)
df.set_index('Country', inplace=True)
```

```python
years = list(map(str, range(1980, 2014)))
df_total = pd.DataFrame(df[years].sum())
df_total.reset_index(inplace=True)
df_total.columns = ['year', 'total']
```

```python
df_total
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1,2, figsize=(16,8))

axes[0].plot(years, df_total['total'])
axes[0].set_title("Immigrants between 1980 to 2013")
axes[0].set_xlabel('Years')
axes[0].set_ylabel('Total Immigrants')


axes[1].scatter(years, df_total['total'])
axes[1].set_title("Immigrants between 1980 to 2013")
axes[1].set_xlabel('Years')
axes[1].set_ylabel('Total Immigrants')
```
