```python
import pandas as pd
import dash
from dash import html, dcc
from dash.dependencies import Input, Output
import plotly.express as px

df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv")
df.isna().sum() #to check for missing values

df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()

app = dash.Dash(__name__)
app.config.suppress_callback_exceptions = True

app.layout = html.Div(children=[
    # TASK 2.1: Title
    html.H1("Automobile Sales Dashboard",
            style={'textAlign': 'center', 'color': 'darkblue', 'font-size': 30}),

    html.Div([

        # TASK 2.2: Region + Year æŽ§åˆ¶å…ƒä»¶
        html.Div([
            html.H2("Select Vehicle Type:", style={'margin-right': '2em'}),
            dcc.RadioItems(
                id='vehicle-choice',
                options=[{'label': v, 'value': v} for v in df['Vehicle_Type'].unique()],
                value=df['Vehicle_Type'].unique()[0],
                inline=True
            )
        ], style={'margin': '20px'}),

        html.Div([
            html.H2("Select Year:", style={'margin-right': '2em'}),
            dcc.Dropdown(
                id='year-choice',
                options=[{'label': y, 'value': y} for y in sorted(df['Year'].unique())],
                value=df['Year'].min()
            )
        ], style={'width': '40%', 'margin': '20px'}),

        # TASK 2.3: Output å®¹å™¨
        html.Div([
            html.Div(id='output-recession', className='six columns'),
            html.Div(id='output-yearly', className='six columns')
        ], style={'display': 'flex', 'flex-direction': 'row'})
    ])
])

# TASK 2.4
@app.callback(
    [Output('output-recession', 'children'),
     Output('output-yearly', 'children')],
    [Input('vehicle-choice', 'value'),
     Input('year-choice', 'value')]
)
def update_graphs(vehicle_type, year_choice):

    df_vehicle = df[df['Vehicle_Type'] == vehicle_type]
    df_year = df[df['Year'] == year_choice]

    # TASK 2.5: Recession Report
    recession_data = df_vehicle[df_vehicle['Recession'] == 1]
    rec_group = recession_data.groupby('Month')['Automobile_Sales'].mean().reset_index()

    fig1 = px.pie(
        rec_group,
        names='Month',
        values='Automobile_Sales',
        title=f"{vehicle_type}: Monthly Avg Sales during Recession"
    )

    # TASK 2.6: Yearly Report
    yearly_group = df_year.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()

    fig2 = px.bar(
        yearly_group,
        x='Vehicle_Type',
        y='Automobile_Sales',
        title=f"Average Sales per Vehicle Type in {year_choice}"
    )

    return [dcc.Graph(figure=fig1), dcc.Graph(figure=fig2)]


if __name__ == '__main__':
    app.run(debug=True, port=8051)

```

## Task 2.4 - Creating Callbacks; Define the callback function to update the input container based on the selected statistics and the output container.

```python
import pandas as pd
import dash
from dash import html, dcc
from dash.dependencies import Input, Output, State, ALL
import plotly.express as px

# è¼‰å…¥è³‡æ–™
df = pd.read_csv(
    "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/"
    "IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
)

# æ¸…ç†ç¼ºå¤±å€¼
df = df.dropna()

# æ–°å¢žæ™‚é–“æ¬„ä½
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()

# å»ºç«‹ Dash app
app = dash.Dash(__name__)

# ========== TASK 2.1: Create a Dash application and give it a meaningful title (2 points) ==========
app.layout = html.Div(children=[
    # Title
    html.H1("Automobile Sales Dashboard",
            style={'textAlign': 'center', 'color': 'darkblue', 'font-size': 30}),
    
    # é¸æ“‡å ±è¡¨é¡žåž‹
    html.Div([
        html.H2("Select Report Type:"),
        dcc.RadioItems(
            id='report-type',
            options=[
                {'label': 'Recession Report', 'value': 'Recession'},
                {'label': 'Yearly Report', 'value': 'Yearly'}
            ],
            value='Recession',
            inline=True
        )
    ], style={'margin': '20px'}),
    
    # TASK 2.2: Add drop-downs to your dashboard with appropriate titles and options (1 point)
    # Recession Report Section
    html.Div(id='recession-section', children=[
        html.H3("Select Year:"),
        dcc.Dropdown(
            id='year-dropdown',
            options=[{'label': y, 'value': y} for y in sorted(df['Year'].unique())],
            value=df['Year'].min(),
            style={'width': '50%'}
        )
    ], style={'margin': '20px', 'display': 'block'}),
    
    # Yearly Report Section  
    html.Div(id='yearly-section', children=[
        html.H3("Select Vehicle Type:"),
        dcc.Dropdown(
            id='vehicle-dropdown',
            options=[{'label': v, 'value': v} for v in df['Vehicle_Type'].unique()],
            value=df['Vehicle_Type'].unique()[0],
            style={'width': '50%'}
        )
    ], style={'margin': '20px', 'display': 'none'}),
    
    # TASK 2.3: Add a division for output display with appropriate 'id' and 'classname' property (1 point)
    html.Div(id='output-container', style={'margin': '20px'})
])

# ========== TASK 2.4: Creating Callbacks; Define the callback function to update the input container (5 points) ==========
@app.callback(
    [Output('recession-section', 'style'),
     Output('yearly-section', 'style')],
    Input('report-type', 'value')
)
def toggle_sections(report_type):
    if report_type == 'Recession':
        return {'margin': '20px', 'display': 'block'}, {'margin': '20px', 'display': 'none'}
    else:
        return {'margin': '20px', 'display': 'none'}, {'margin': '20px', 'display': 'block'}

# ========== Callback 2: æ›´æ–°åœ–è¡¨ ==========
@app.callback(
    Output('output-container', 'children'),
    [Input('report-type', 'value'),
     Input('year-dropdown', 'value'),
     Input('vehicle-dropdown', 'value')]
)
def update_charts(report_type, selected_year, selected_vehicle):
    
    # ========== TASK 2.5: Create and display graphs for Recession Report Statistics (3 points) ==========
    if report_type == 'Recession':
        if selected_year is None:
            return html.Div("Please select a year.")
        
        # ç¯©é¸è¡°é€€æœŸè³‡æ–™
        df_recession = df[(df['Recession'] == 1) & (df['Year'] == selected_year)]
        
        if df_recession.empty:
            return html.Div(f"No recession data available for year {selected_year}.")
        
        # æœˆä»½é †åº
        month_order = ['January', 'February', 'March', 'April', 'May', 'June',
                       'July', 'August', 'September', 'October', 'November', 'December']
        
        # æŒ‰æœˆåˆ†çµ„è¨ˆç®—å¹³å‡éŠ·å”®é‡
        sales_group = df_recession.groupby('Month')['Automobile_Sales'].mean().reset_index()
        sales_group['Month'] = pd.Categorical(sales_group['Month'], categories=month_order, ordered=True)
        sales_group = sales_group.sort_values('Month').reset_index(drop=True)
        
        # æŒ‰æœˆåˆ†çµ„è¨ˆç®—å¹³å‡å»£å‘Šæ”¯å‡º
        ad_group = df_recession.groupby('Month')['Advertising_Expenditure'].mean().reset_index()
        ad_group['Month'] = pd.Categorical(ad_group['Month'], categories=month_order, ordered=True)
        ad_group = ad_group.sort_values('Month').reset_index(drop=True)
        
        # å»ºç«‹åœ–è¡¨
        fig1 = px.line(sales_group, x='Month', y='Automobile_Sales',
                       title=f"Monthly Average Sales during Recession ({selected_year})",
                       markers=True)
        fig1.update_layout(
            xaxis_title="Month", 
            yaxis_title="Average Sales",
            height=400
        )
        
        fig2 = px.bar(ad_group, x='Month', y='Advertising_Expenditure',
                      title=f"Monthly Average Advertising Expenditure during Recession ({selected_year})")
        fig2.update_layout(
            xaxis_title="Month", 
            yaxis_title="Average Advertising Expenditure",
            height=400
        )
        
        return html.Div([
            dcc.Graph(figure=fig1),
            html.Br(),
            dcc.Graph(figure=fig2)
        ])
    
    # ========== TASK 2.6: Create and display graphs for Yearly Report Statistics (2 points) ==========
    elif report_type == 'Yearly':
        if selected_vehicle is None:
            return html.Div("Please select a vehicle type.")
        
        # ç¯©é¸è»Šåž‹è³‡æ–™
        df_vehicle = df[df['Vehicle_Type'] == selected_vehicle]
        
        if df_vehicle.empty:
            return html.Div(f"No data available for vehicle type: {selected_vehicle}")
        
        # æŒ‰å¹´è¨ˆç®—å¹³å‡éŠ·å”®é‡
        yearly_sales = df_vehicle.groupby('Year')['Automobile_Sales'].mean().reset_index()
        yearly_sales = yearly_sales.sort_values('Year').reset_index(drop=True)
        
        # å»ºç«‹åœ–è¡¨
        fig = px.bar(yearly_sales, x='Year', y='Automobile_Sales',-
                     title=f"Yearly Average Sales for {selected_vehicle}")
        fig.update_layout(
            xaxis_title="Year", 
            yaxis_title="Average Sales",
            height=500
        )
        
        return dcc.Graph(figure=fig)
    
    return html.Div("Please select report type and input options.")

# ä¸»ç¨‹å¼å…¥å£
if __name__ == '__main__':
    app.run(debug=True, port=8051)
```

## Task 2.5 - Create and display graphs for Recession Report Statistics.

```python
import pandas as pd
import dash
from dash import html, dcc
from dash.dependencies import Input, Output
import plotly.express as px

# è¼‰å…¥è³‡æ–™ (è«‹ç¢ºèªæœ¬åœ°æœ‰é€™å€‹æª”æ¡ˆ Automobile_sales.csv)
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv")

# é è™•ç†
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()

# å»ºç«‹ Dash app
app = dash.Dash(__name__)
app.config.suppress_callback_exceptions = True

# ===== Layout =====
app.layout = html.Div(children=[
    # TASK 2.1: Title
    html.H1("Automobile Sales Dashboard",
            style={'textAlign': 'center', 'color': 'darkblue', 'font-size': 30}),

    # é¸æ“‡å ±è¡¨é¡žåž‹
    html.Div([
        html.H2("Select Report Type:"),
        dcc.RadioItems(
            id='report-type',
            options=[
                {'label': 'Recession Report', 'value': 'Recession'},
                {'label': 'Yearly Report', 'value': 'Yearly'}
            ],
            value='Recession',
            inline=True
        )
    ], style={'margin': '20px'}),

    # Task 2.2: Input container (æœƒç”± callback å‹•æ…‹æ›´æ–°)
    html.Div(id='input-container', style={'margin': '20px'}),

    # Task 2.3: Output container
    html.Div(id='output-container')
])

# ===== Callback 1: æ›´æ–° Input Container =====
@app.callback(
    Output('input-container', 'children'),
    Input('report-type', 'value')
)
def update_input_container(selected_report):
    if selected_report == 'Recession':
        return html.Div([
            html.H3("Select Year for Recession Report:"),
            dcc.Dropdown(
                id='recession-year',
                options=[{'label': y, 'value': y} for y in sorted(df['Year'].unique())],
                value=df['Year'].min()
            )
        ])
    elif selected_report == 'Yearly':
        return html.Div([
            html.H3("Select Vehicle Type for Yearly Report:"),
            dcc.Dropdown(
                id='yearly-vehicle',
                options=[{'label': v, 'value': v} for v in df['Vehicle_Type'].unique()],
                value=df['Vehicle_Type'].unique()[0]
            )
        ])

# ===== Callback 2: æ›´æ–° Output Container =====
from dash import no_update

@app.callback(
    Output('output-container', 'children'),
    [Input('report-type', 'value'),
     Input('recession-year', 'value'),
     Input('yearly-vehicle', 'value')]
)
def update_output(report_type, year, vehicle):
    # Recession Report
    if report_type == 'Recession' and year is not None:
        df_recession = df[(df['Recession'] == 1) & (df['Year'] == year)]
        sales_data = df_recession.groupby('Month')['Automobile_Sales'].mean().reset_index()
        ad_data = df_recession.groupby('Month')['Advertising_Expenditure'].mean().reset_index()

        fig1 = px.line(sales_data, x='Month', y='Automobile_Sales',
                       title=f"Monthly Average Sales during Recession ({year})")
        fig2 = px.bar(ad_data, x='Month', y='Advertising_Expenditure',
                      title=f"Monthly Average Ad Expenditure during Recession ({year})")

        return html.Div([
            dcc.Graph(figure=fig1),
            dcc.Graph(figure=fig2)
        ])

    # Yearly Report
    elif report_type == 'Yearly' and vehicle is not None:
        df_vehicle = df[df['Vehicle_Type'] == vehicle]
        yearly_sales = df_vehicle.groupby('Year')['Automobile_Sales'].mean().reset_index()

        fig = px.bar(yearly_sales, x='Year', y='Automobile_Sales',
                     title=f"Yearly Average Sales for {vehicle}")

        return dcc.Graph(figure=fig)

    # é è¨­ï¼šä¸è¦å ±éŒ¯ï¼Œå›žå‚³ç©º
    return no_update


# ä¸»ç¨‹å¼å…¥å£
if __name__ == '__main__':
    app.run(debug=True, port=8051)
```
