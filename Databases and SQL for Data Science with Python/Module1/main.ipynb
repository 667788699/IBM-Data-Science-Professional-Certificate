## introduction to Databases


---
###  1. SQL (Structured Query Language)

* 一種用於操作與查詢資料庫的語言。

###  2. Data（資料）

* 原始事實或統計值，可被儲存與分析。

###  3. Database（資料庫）

* 有組織地儲存資料的集合，便於查詢與管理。

###  4. Relational Databases（關聯式資料庫）

* 以表格（Table）方式儲存資料，各表之間透過鍵（Key）建立關聯。

###  5. RDBMS（Relational Database Management System）

* 關聯式資料庫管理系統，用來建立、管理和操作關聯式資料庫的軟體，如 MySQL、PostgreSQL、SQL Server。

###  6. 5 個基本 SQL 指令

* `CREATE`：建立資料表或資料庫
* `INSERT`：新增資料
* `SELECT`：查詢資料
* `UPDATE`：更新資料
* `DELETE`：刪除資料

##  主題：使用 SELECT 語句查詢資料（Retrieving Data with SELECT Statement）

###  課程目標：

學習如何從關聯式資料表中查詢資料，包含以下技能：

* 查詢表格中的資料（retrieve data from a table）
* 使用條件語句（predicate）來篩選資料
* 寫出包含 `WHERE` 子句的 `SELECT` 語法
* 認識 RDBMS 支援的比較運算子（comparison operators）

---

### 內容摘要：

####  SELECT 是什麼？

* `SELECT` 是 **資料操作語言（DML）** 的一部分。
* 用來從資料表中查詢資料。
* 查詢的結果稱為 **結果集（result set）** 或 **結果表（result table）**。

####  基本語法：

```sql
SELECT * FROM table_name;
```

* `*` 代表選取該表格的 **所有欄位**。

####  範例：

* 假設有一個名為 `book` 的資料表。
* 使用：

  ```sql
  SELECT * FROM book;
  ```

  可查詢該表中 **所有欄位的所有資料列（rows）**。

####  查詢部分欄位：

* 若只想查 `book_id` 與 `title` 兩個欄位：

  ```sql
  SELECT book_id, title FROM book;
  ```
* 顯示順序會與 `SELECT` 中欄位的順序一致。

####  使用 WHERE 子句篩選資料：

* 若只查詢 `book_id = 'B1'` 的書籍標題：

  ```sql
  SELECT book_id, title FROM book WHERE book_id = 'B1';
  ```
* `WHERE` 子句中使用的是 **謂詞（predicate）**，這是一個會產生 **True / False / Unknown** 結果的條件。

####  常見比較運算子：

| 運算子     | 意義    |
| ------- | ----- |
| =       | 等於    |
| >       | 大於    |
| <       | 小於    |
| >=      | 大於或等於 |
| <=      | 小於或等於 |
| <> 或 != | 不等於   |
