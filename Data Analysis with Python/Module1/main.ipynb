# Understanding the Data


## use pandas read cvs (url)

```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

df = pd.read_csv(url, header=None) # not assign header
df.head()
```

```python
df.tail() # last five of the data
```

## sign names to df

```python
column_names = [
    "symboling", "normalized-losses", "make", "fuel-type", "aspiration",
    "num-of-doors", "body-style", "drive-wheels", "engine-location",
    "wheel-base", "length", "width", "height", "curb-weight",
    "engine-type", "num-of-cylinders", "engine-size", "fuel-system",
    "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm",
    "city-mpg", "highway-mpg", "price"
]

df = pd.read_csv(url, names=column_names)
df
```

## exporting a Pandas dataframe to CSV


```python
filePath = "Analysis/Data/imports-85.csv"
df.to_csv(filePath)
```

## Q1 check the bottom 10 rows of dataframe "df"

```python
import pandas as pd 

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(url)
df.columns = headers
df.tail(10)
```

## replace the ?

```python
import numpy as np
df.replace('?', np.NAN)
```

## You can drop missing values along the column "price" as follows:

```python
df_features = df.dropna(subset=["price"], axis=0)
df_features
```

```python
df_features = df.drop(columns='price')
df_features
```
