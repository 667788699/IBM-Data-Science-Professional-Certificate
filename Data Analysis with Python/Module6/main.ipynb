## Q8

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


X = df[["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]]
y = df['price']


pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])

pipe.fit(X, y)

y_pred = pipe.predict(X)

r2 = r2_score(y, y_pred)
print("R^2:", r2)
```

## Q9

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)

print("number of training samples:",x_train.shape[0])
print("number of test samples:", x_test.shape[0])
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

ridge = Ridge(alpha=0.1)

ridge.fit(x_train, y_train)

yPredTrain = ridge.predict(x_train)

r2 = r2_score(y_train, yPredTrain)
print(f"R2-score of Training : {r2}")
```

## Q10

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

pr = PolynomialFeatures(degree=2)

xTrain_pr = pr.fit_transform(x_train)
xTest_pr = pr.fit_transform(x_test)

poly = Ridge(alpha=0.1)

poly.fit(xTrain_pr, y_train)

yPredTest = poly.predict(xTest_pr)
r2Test = r2_score(y_test, yPredTest)

print(f"R2-score of Test data with polynoninal features : {r2Test}")

```
