# Exploratory Data Analysis

## Descriptive Statistics

```python
import pandas as pd
import os


url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

column_names = [
    "symboling", "normalized-losses", "make", "fuel-type", "aspiration",
    "num-of-doors", "body-style", "drive-wheels", "engine-location",
    "wheel-base", "length", "width", "height", "curb-weight",
    "engine-type", "num-of-cylinders", "engine-size", "fuel-system",
    "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm",
    "city-mpg", "highway-mpg", "price"
]

os.makedirs("Analysis/Data", exist_ok=True)
df = pd.read_csv(url,names=column_names, na_values="?")
df.dropna(inplace=True)
df.to_csv("Analysis/Data/automatic.csv")
df
```

```python
df.describe()
```

```python
df.value_counts()
```

```python
import matplotlib.pyplot as plt 
x = df["horsepower"]
y = df["price"]

plt.figure(figsize=(8,6))
plt.scatter(x, y)
plt.title("Scatter Plot of Horsepower vs. Price")
plt.xlabel("HorsePower")
plt.ylabel("Price")
```

## GroupBy in Python

```python
print(df["drive-wheels"].value_counts())
print(df["body-style"].value_counts())
```

```python
df_test = df[["drive-wheels", "body-style", "price"]]

df_group = df_test.groupby(["drive-wheels", "body-style"], as_index=False).mean()
df_group
```

```python
import numpy as np
df_pivot = df_group.pivot(index="drive-wheels", columns="body-style")
df_pivot.replace(np.nan, 0, inplace=True)
df_pivot
```

### Heatmap

```python
import matplotlib.pyplot as plt 

plt.pcolor(df_pivot, cmap='RdBu')
plt.colorbar()
plt.xticks(np.arange(0.5, len(df_pivot.columns), 1), df_pivot.columns, rotation=45)
plt.yticks(np.arange(0.5, len(df_pivot.index), 1), df_pivot.index)
plt.xlabel("body-style")
plt.ylabel("drive-wheels")
plt.title("Average Price by Drive Wheels and Body Style")

plt.show()
```

## Correlation - Statistics


```python
import seaborn as sns
import matplotlib.pyplot as plt

# åªå–æ•¸å€¼æ¬„ä½
corr_matrix = df.select_dtypes(include=['number']).corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Pearson Correlation Coefficient (PCC) Heatmap")
plt.show()
```

```python
from scipy import stats

p_coef, p_value = stats.pearsonr(df["horsepower"], df["price"])
print(f"{p_coef} and {p_value}")
```

```python
df_q2 = df[["bore", "stroke", "compression-ratio", "horsepower"]]
corrMat = df_q2.corr()

import seaborn as sns

sns.heatmap(corrMat)
```
