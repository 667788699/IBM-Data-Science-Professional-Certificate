## Bar Chart of Success Rate per Orbit Type

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import requests
import io

# --- 1. Download Data ---
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"

response = requests.get(URL)
response.raise_for_status()
df = pd.read_csv(io.BytesIO(response.content))

# --- 2. Compute success rate per orbit ---
# Class = 1 (success), 0 (failure)
orbit_success_rate = df.groupby("Orbit")["Class"].mean().sort_values()

# --- 3. Bar chart ---
plt.figure(figsize=(12, 6))

sns.barplot(
    x=orbit_success_rate.index,
    y=orbit_success_rate.values,
    palette="deep"
)

plt.title("Success Rate for Each Orbit Type", fontsize=16)
plt.xlabel("Orbit Type", fontsize=14)
plt.ylabel("Success Rate", fontsize=14)

plt.xticks(rotation=45, ha="right")
plt.ylim(0, 1)

plt.show()

# --- 4. Explanation for screenshot ---
print("""
Bar Chart Explanation:
- Each bar represents the launch success rate for a specific orbit type.
- Success rate is calculated as: (successful launches) / (total launches) for that orbit.
- Some orbit types (e.g., GTO or LEO variants) show high launch success consistency.
- Others with fewer launch attempts may show more variability in success rate.
- This visualization helps identify which orbit targets historically have more reliable launch outcomes.
""")
```

# Launch Success Yearly Trend


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import requests
import io

# --- 1. Download Data ---
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"

response = requests.get(URL)
response.raise_for_status()
df = pd.read_csv(io.BytesIO(response.content))

# --- 2. Extract Year from Date ---
df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year

# --- 3. Compute YEARLY success rate ---
yearly_success = df.groupby("Year")["Class"].mean()

# --- 4. Line Chart ---
plt.figure(figsize=(12, 6))

sns.lineplot(
    x=yearly_success.index,
    y=yearly_success.values,
    marker="o",
    linewidth=2
)

plt.title("Yearly Average Launch Success Rate", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Success Rate", fontsize=14)
plt.ylim(0, 1)

plt.grid(True, linestyle="--", alpha=0.4)

plt.show()

# --- 5. Explanation for screenshot ---
print("""
Line Chart Explanation:
- The line chart shows the average Falcon 9 launch success rate for each year.
- Success rate is calculated as: (number of successful launches) / (total launches) for that year.
- The upward trend across the years reflects SpaceXâs increasing reliability and technological improvements.
- Early years show lower stability, while later years approach near-perfect success.
""")
```

# Successful Drone Ship Landing with Payload between 4000 and 6000


```python
import pandas as pd

# Load dataset
df = pd.read_csv("spacex_launch_geo.csv")

# Filter conditions:
# 1. Successful landing on drone ship
# 2. Payload mass between 4000 and 6000 kg
filtered = df[
    df["Landing Outcome"].str.contains("Success (drone ship)", regex=False, na=False) &
    (df["Payload Mass (kg)"] > 4000) &
    (df["Payload Mass (kg)"] < 6000)
]

# Select booster names and drop duplicates
result = filtered["Booster Version"].drop_duplicates()

# Display result
print(result)

# Explanation
print("\nExplanation:")
print("We filtered the dataset to include only launches with successful drone ship")
print("landings and with payload mass between 4000 and 6000 kg. Then we extracted")
print("the unique booster versions that meet these criteria.")
```

# Rank Landing Outcomes Between 2010-06-04 and 2017-03-20


```python
import pandas as pd

# Load dataset
df = pd.read_csv("spacex_launch_geo.csv")

# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Filter for date range
mask = (df["Date"] >= "2010-06-04") & (df["Date"] <= "2017-03-20")
filtered_df = df[mask]

# Count and sort landing outcomes
landing_rank = (
    filtered_df["Landing Outcome"]
    .value_counts()
    .sort_values(ascending=False)
)

# Display result
print("Landing Outcome Ranking (2010-06-04 to 2017-03-20):")
print(landing_rank)

# Explanation
print("\nExplanation:")
print("We filtered launches within the specified date range, then counted how many")
print("times each landing outcome occurred. Finally, we sorted")
```

# <Folium Map Screenshot 2>

```python
import folium
import pandas as pd

# Load dataset
df = pd.read_csv("spacex_launch_geo.csv")

# Simple color mapping for landing outcomes
def color_producer(outcome):
    if "Success" in outcome:
        return "green"
    elif "Failure" in outcome:
        return "red"
    elif "No attempt" in outcome:
        return "gray"
    else:
        return "blue"

# Base map (global view to show all sites & outcomes)
m = folium.Map(location=[20, 0], zoom_start=2)

# Add markers using launch outcome color coding
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row["Lat"], row["Long"]],
        radius=6,
        color=color_producer(row["Landing Outcome"]),
        fill=True,
        fill_color=color_producer(row["Landing Outcome"]),
        fill_opacity=0.8,
        popup=f"Site: {row['Launch Site']}<br>Outcome: {row['Landing Outcome']}"
    ).add_to(m)

# Display map
m
```

# <Folium Map Screenshot 3>

```python
import folium
from geopy.distance import geodesic

# --- Launch Site Coordinates ---
launch_sites = {
    "CCAFS SLC-40": (28.561857, -80.577366),
}

# Select one site
site_name = "CCAFS SLC-40"
site_coord = launch_sites[site_name]

# --- Nearby Features (example coordinates) ---
railway_coord = (28.5622, -80.5900)
highway_coord = (28.5638, -80.5700)
coastline_coord = (28.5600, -80.5600)

# --- Create Map ---
m = folium.Map(location=site_coord, zoom_start=14)

# Launch Site Marker
folium.Marker(
    location=site_coord,
    icon=folium.Icon(color="red", icon="rocket"),
    popup=f"{site_name}"
).add_to(m)

# Feature Markers
folium.Marker(railway_coord, tooltip="Railway", icon=folium.Icon(color="blue")).add_to(m)
folium.Marker(highway_coord, tooltip="Highway", icon=folium.Icon(color="green")).add_to(m)
folium.Marker(coastline_coord, tooltip="Coastline", icon=folium.Icon(color="purple")).add_to(m)

# --- Distance Calculation ---
dist_rail = geodesic(site_coord, railway_coord).km
dist_highway = geodesic(site_coord, highway_coord).km
dist_coast = geodesic(site_coord, coastline_coord).km

# --- Draw lines and labels ---
folium.PolyLine([site_coord, railway_coord], color="blue").add_to(m)
folium.PolyLine([site_coord, highway_coord], color="green").add_to(m)
folium.PolyLine([site_coord, coastline_coord], color="purple").add_to(m)

# Distance labels
folium.Marker(
    [(site_coord[0] + railway_coord[0]) / 2, (site_coord[1] + railway_coord[1]) / 2],
    tooltip=f"To Railway: {dist_rail:.2f} km"
).add_to(m)

folium.Marker(
    [(site_coord[0] + highway_coord[0]) / 2, (site_coord[1] + highway_coord[1]) / 2],
    tooltip=f"To Highway: {dist_highway:.2f} km"
).add_to(m)

folium.Marker(
    [(site_coord[0] + coastline_coord[0]) / 2, (site_coord[1] + coastline_coord[1]) / 2],
    tooltip=f"To Coastline: {dist_coast:.2f} km"
).add_to(m)

# Display map
m
```

```python
import pandas as pd
import os
import requests

csv_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_dash.csv"
csv_file = "spacex_launch_dash.csv"

# If CSV doesn't exist, download it
if not os.path.exists(csv_file):
    print("Downloading spacex_launch_dash.csv ...")
    r = requests.get(csv_url)
    with open(csv_file, "wb") as f:
        f.write(r.content)
    print("Download complete!")

# Load dataset
spacex_df = pd.read_csv(csv_file)

# Show first rows to confirm successful load
print(spacex_df.head())
```

```python
# Import required libraries
import pandas as pd
import dash
from dash import html
from dash import dcc
from dash.dependencies import Input, Output
import plotly.express as px

# Read the SpaceX Launch dataset
spacex_df = pd.read_csv("spacex_launch_dash.csv")

# Get the maximum and minimum payload mass
max_payload = spacex_df["Payload Mass (kg)"].max()
min_payload = spacex_df["Payload Mass (kg)"].min()

# Create a Dash application
app = dash.Dash(__name__)

# Application layout
app.layout = html.Div(children=[
    html.H1("SpaceX Launch Records Dashboard", 
            style={'textAlign': 'center', 'color': '#503D36', 'font-size': 40}),

    # TASK 1: Launch Site Dropdown
    dcc.Dropdown(
        id='site-dropdown',
        options=[
            {'label': 'All Sites', 'value': 'ALL'},
            *[
                {'label': site, 'value': site}
                for site in spacex_df["Launch Site"].unique()
            ]
        ],
        value='ALL',
        placeholder="Select a Launch Site",
        searchable=True
    ),

    html.Br(),

    # TASK 2: Pie Chart Output
    html.Div(dcc.Graph(id='success-pie-chart')),
    html.Br(),

    # TASK 3: Payload Range Slider
    html.P("Payload range (Kg):"),
    dcc.RangeSlider(
        id='payload-slider',
        min=0, max=10000, step=1000,
        marks={0: '0', 2500: '2500', 5000: '5000', 7500: '7500', 10000: '10000'},
        value=[min_payload, max_payload]
    ),

    # TASK 4: Scatter Plot Output
    html.Div(dcc.Graph(id='success-payload-scatter-chart')),
])

# TASK 2 Callback: Pie Chart Logic
@app.callback(
    Output('success-pie-chart', 'figure'),
    Input('site-dropdown', 'value'))
def update_pie_chart(selected_site):
    if selected_site == "ALL":
        fig = px.pie(
            spacex_df,
            values='class',
            names='Launch Site',
            title='Total Success Launches for All Sites'
        )
        return fig
    else:
        filtered_df = spacex_df[spacex_df["Launch Site"] == selected_site]
        fig = px.pie(
            filtered_df,
            names='class',
            title=f"Success vs Failure for {selected_site}"
        )
        return fig

# TASK 4 Callback: Scatter Plot Logic
@app.callback(
    Output('success-payload-scatter-chart', 'figure'),
    [
        Input('site-dropdown', 'value'),
        Input('payload-slider', 'value')
    ]
)
def update_scatter(selected_site, payload_range):
    low, high = payload_range
    filtered_df = spacex_df[
        (spacex_df["Payload Mass (kg)"] >= low) &
        (spacex_df["Payload Mass (kg)"] <= high)
    ]

    if selected_site != "ALL":
        filtered_df = filtered_df[filtered_df["Launch Site"] == selected_site]

    fig = px.scatter(
        filtered_df,
        x="Payload Mass (kg)",
        y="class",
        color="Booster Version Category",
        title="Payload vs Launch Outcome"
    )
    return fig

# Run the application
if __name__ == '__main__':
    app.run_server()
```
